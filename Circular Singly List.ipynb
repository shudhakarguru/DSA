```python
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None
class CircularLinkedList:
    def __init__(self):
        self.head=None
    #traversal method
    def print_LL(self):
        if self.head is None:
            print('Linked List is empty')
        else:
            n=self.head
            while n:
                print(n.data,'-->',end=' ')
                if n.ref==self.head:
                    break
                n=n.ref
    #inserting element at the beginning
    def add_begin(self,data):
        if self.head is None:
            newNode = Node(data)
            self.head = newNode
            newNode.ref = self.head
        elif self.head == self.head.ref:
            newNode = Node(data)
            self.head.ref = newNode
            newNode.ref = self.head
            self.head = newNode
        else:   
            n=self.head
            newNode = Node(data)
            newNode.ref = self.head
            while n.ref!=self.head:
                n=n.ref
            self.head = newNode
            n.ref = newNode
    #inserting at the end
    def add_end(self,data):
        newNode=Node(data)
        if self.head is None:
            self.head = newNode
            newNode.ref = self.head
        else:
            n=self.head
            while n.ref!=self.head:
                n=n.ref
            n.ref=newNode
        newNode.ref = self.head
    #add after a value
    def add_after(self,data,x):
        if self.head is None:
            print('SLL is empty')
        else:
            n=self.head
            while n.ref!=self.head:
                if n.data == x:
                    break
                else:
                    n=n.ref
            if n.data==x:
                newNode = Node(data)
                newNode.ref = n.ref
                n.ref = newNode
            else:
                print("value not in SLL")
                
    #add before a value
    def add_before(self,data,x):
        if self.head is None:
            print('SLL is empty')
        else:
            n=self.head
            if self.head.data == x:
                newNode = Node(data)
                newNode.ref = self.head
                while n.ref!=self.head:
                    n=n.ref
                self.head = newNode
                n.ref = newNode
            else:
                while n.ref!=self.head:
                    if n.ref.data == x:
                        break
                    else:
                        n=n.ref
                if n.ref.data==x:
                    newNode = Node(data)
                    newNode.ref = n.ref
                    n.ref = newNode
                else:
                    print('Node is not found')
                    
                    
    # delete from beginning
    def delete_begin(self):
        if self.head is None:
            print('SLL is empty')
        else:
            n=self.head
            while n:
                if n.ref==self.head:
                    break
                n=n.ref
            self.head = self.head.ref
            n.ref = self.head
            
    # delete from end
    def delete_end(self):
        if self.head is None:
            print('SLL is empty')
        else:
            if self.head.ref==self.head:
                self.head = None
            else:
                n=self.head
                while n.ref.ref!=self.head:
                    n=n.ref
                n.ref = self.head
                
    # delete by value
    def delete_by_value(self,x):
        if self.head is None:
            print('SLL is empty')
        else:
            if self.head.data == x:
                if self.head.ref==self.head:
                    self.head = None
                else:
                    n=self.head
                    self.head = self.head.ref
                    while n:
                        if n.ref.ref==self.head:
                            break
                        n=n.ref
                    n.ref = self.head
            else:
                n=self.head
                while n.ref!=self.head:
                    if x==n.ref.data:
                        break
                    else:
                        n=n.ref
                if n.ref==self.head:
                    print('value not found')
                else:
                    n.ref = n.ref.ref
                    
```
